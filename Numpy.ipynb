Broadcasting - we will only able to perform any mathematical function on two numpy arrays if they have equal number of elements or any one should have only single number

```python
x= np.array ([1,2,3])
v = np.array([4,5,6])
```

```python
x+v     # ---This function is works fine since two numpy array have equal number of elements
```

```python
b = np.array([1,4,8])
t = np.array([3,5])
```

```python
b + t   # ------This will not work since it dont have equal number of elements in numpy array
```

```python
w= np.array([25,45,69])
u = np.array([2])
```

```python
w+u   #----This will work since one of numpy array have single number - this is where broadcasting concept comes into picture
```
